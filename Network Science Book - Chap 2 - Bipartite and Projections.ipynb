```python
import networkx as nx
import matplotlib.pyplot as plt
```

# Bipartite Graph
###### Albeit useful to model systems with 2 sets of 'objects' that only interact with objects from a different set, bipartite graphs seems to be 'rare'
###### Exemple of a bipartite graph: SetA = 'Movies', SetB='people', System = 'Netflix'

```python
#Replicating the same graph presented in Image 2.9
U_set=[1,2,3,4,5,6,7] # coulb be people
V_set=['A','B','C','D'] # could de movies
G = nx.Graph()
G.add_nodes_from(U_set,bipartite=0)
G.add_nodes_from(V_set,bipartite=1)
#Person 1 whatched Movie 'A'
#Person 2 watched Movie 'A' and 'B'
G.add_edges_from([
    (1,'A'),
    (2,'A'),
    (2,'B'),
    (3,'A'),
    (4,'C'),
    (5,'B'),
    (5,'C'),
    (5,'D'),
    (6,'D'),
    (7,'D')
])
print('drawing like a "normal" graph')
plt.figure(figsize=(8,8))
nx.draw_kamada_kawai(G,with_labels=1,node_size=700)
plt.show()

print('drawing as a bipartite graph')
plt.figure(figsize=(10,10))
pos = nx.bipartite_layout(G,U_set)
nx.draw_networkx(G,pos=pos,nodelist=U_set,node_color='green',node_size=1000)
nx.draw_networkx(G,pos=pos,nodelist=V_set,node_color='purple',node_size=3000)
#nx.draw(G,pos=pos,with_labels=1)
plt.tight_layout()
plt.show()
```

# Projections
###### Connect nodes in the same set considering they relation with the other set

```python
plt.figure(figsize=(10,10))
P1 = nx.project(G,U_set)
nx.draw_kamada_kawai(P1,with_labels=1,node_color='green',node_size=1500)
plt.show()

plt.figure(figsize=(10,10))
P2 = nx.project(G,V_set)
nx.draw_kamada_kawai(P2,with_labels=1,node_color='purple',node_size=1500)
plt.show()
```
