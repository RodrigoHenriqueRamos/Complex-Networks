# Ploting Networks
######  Even though NetworkX's main purpose is to analyze the networks and not draw them,  It's useful to visualize small networks direct on Jupyter.
###### For visualization of bigger networks, Gephi and Cytoscape are recommended

```python
import networkx as nx
import matplotlib.pyplot as plt
```

# Networks used for exemplification

```python
G1 = nx.balanced_tree(2,5) ; G1.name = "Binary Tree"
G2 = nx.tutte_graph() ; G2.name = "Tute"
G3 = nx.complete_graph(12) ; G3.name = "Complete 12"
G4 = nx.circular_ladder_graph(10) ; G4.name = "Circular Ladder"
networks = [G1,G2,G3,G4]
for net in networks:
    print(nx.info(net),end='\n\n')
```

# Layouts

## Random

```python
fig, ax = plt.subplots(2,2,figsize=(15, 7))
col=0;line=0
for net in networks:
    plt.sca(ax[line,col])
    ax[line,col] = nx.draw_random(net,with_labels=True, node_size=100, font_size=7)
    col+=1
    if(col==2):
        col=0
        line+=1
plt.show()
```

## Circular

```python
fig, ax = plt.subplots(2,2,figsize=(15, 7))
col=0;line=0
for net in networks:
    plt.sca(ax[line,col])
    ax[line,col] = nx.draw_circular(net,with_labels=True, node_size=100, font_size=7)
    col+=1
    if(col==2):
        col=0
        line+=1
plt.show()
```

## Shell (Same as Circular?)
###### The NetworkX documentation did not mention any similarity between Shell and Circular layout, but they seem the same.

```python
fig, ax = plt.subplots(2,2,figsize=(15, 7))
col=0;line=0
for net in networks:
    plt.sca(ax[line,col])
    ax[line,col] = nx.draw_shell(net,with_labels=True, node_size=100, font_size=7)
    col+=1
    if(col==2):
        col=0
        line+=1
plt.show()
```

## Kamada-Kawai (Personally my favorite)
###### Draw the graph G with a Kamada-Kawai force-directed layout.
###### Kamada-Kawai uses the edge weight as distance. If no weight is given, 1 is assumes

```python
fig, ax = plt.subplots(2,2,figsize=(15, 7))
col=0;line=0
for net in networks:
    plt.sca(ax[line,col])
    ax[line,col] = nx.draw_kamada_kawai(net,with_labels=True, node_size=100, font_size=7)
    col+=1
    if(col==2):
        col=0
        line+=1
plt.show()
```

## Spring
###### Here we fixed the seed to keep the same result.

```python
fig, ax = plt.subplots(2,2,figsize=(15, 7))
col=0;line=0
for net in networks:
    plt.sca(ax[line,col])
    ax[line,col] = nx.draw_spring(net,with_labels=True, node_size=100, font_size=7, seed=44)
    col+=1
    if(col==2):
        col=0
        line+=1
plt.show()
```

## Planar
###### Planar graphs don't intersect edges. For this reason, G3 can't be plotted and is changed for a complete graph of size 4

```python
fig, ax = plt.subplots(2,2,figsize=(15, 7))
plt.sca(ax[0,0])
ax[0,0] = nx.draw_planar(networks[0],with_labels=True, node_size=100, font_size=7)
plt.sca(ax[0,1])
ax[0,1] = nx.draw_planar(networks[1],with_labels=True, node_size=100, font_size=7)
plt.sca(ax[1,0])
ax[1,0] = nx.draw_planar(nx.complete_graph(4),with_labels=True, node_size=100, font_size=7)
plt.sca(ax[1,1])
ax[1,1] = nx.draw_planar(networks[3],with_labels=True, node_size=100, font_size=7)
plt.show()
```

## Spectral
###### Planar graphs don't intersect edges. For this reason, G3 can't be plotted and is changed for a complete graph of size 4

```python
fig, ax = plt.subplots(2,2,figsize=(15, 7))
plt.sca(ax[0,0])
ax[0,0] = nx.draw_spectral(networks[0],with_labels=True, node_size=100, font_size=7)
plt.sca(ax[0,1])
ax[0,1] = nx.draw_spectral(networks[1],with_labels=True, node_size=100, font_size=7)
plt.sca(ax[1,0])
ax[1,0] = nx.draw_spectral(networks[2],with_labels=True, node_size=100, font_size=7)
plt.sca(ax[1,1])
ax[1,1] = nx.draw_spectral(networks[3],with_labels=True, node_size=100, font_size=7)
plt.show()
```
